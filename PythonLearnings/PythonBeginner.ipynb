```python
"""
Given a list of elements i.e[1,2,4,-5,7,9,3,2] create another list with these elements sorted from min to max
"""

L=[1,2,4,-5,7,9,3,2]
n=len(L)
for i in range(n):
    for j in range(0, n - i - 1):
            # Traverse the array from 0 to n - i - 1
            # Swap if the element found is greater than the next element
            if L[j] > L[j + 1]:
                L[j], L[j + 1] = L[j + 1], L[j]
print(L)
    
    
```

```python
"""Given a list of elements i.e[1,2,4,-5,7,9,3,2] create another list with these elements sorted from min to max"""

def sort(*args):
    args_list=list(args)
    if not(checkArgs2(args_list)):
        print("error not numeric")
        return
    else:
        n=len(args_list)
        for i in range(n) :
            for j in range(0,n-i-1):
                if args_list[j]>args_list[j+1]:
                    args_list[j],args_list[j+1]=args_list[j+1],args_list[j]
        return args_list
        
```

```python
"""
Lets say you are teacher and you have different student records containing Id's and marklist of students on which each students have taken different number of subjects.You want to enter all this data and you want to compute the average of the marks of each student and display
"""

def getData():
    D={}
    while True:
        studentId=input("Enter input Id")
        Marks=input("enter marks with comma separated values")
        morestudents=input("Enter \"no\" for insertion")
        if studentId in D:
            print("Student already present")
        else:
            D[studentId]=Marks.split(',')
        if morestudents.lower()=="no":
            return D
```

```python
Data=getData()
```

```python
Data
```

```python
def getAvgMarks(D):
    avgMarks={}
    for x in D:
        L=D[x]
        s=0
        for marks in L:
            s+=int(marks)
        avgMarks[x]=s/len(L)
    return avgMarks
            
    
```

```python
avgMarks=getAvgMarks(Data)
```

```python
avgMarks
```

```python
for x in avgMarks:
    print("student",x,"got average marks of",avgMarks[x])
```

In NumPy, when you create an array with a single scalar value using np.array(3), it creates a zero-dimensional array or a scalar array. This array has no dimensions other than the value itself. It's essentially just a single value without any axes. Therefore, its dimensionality is 0.

However, when you create an array with a single value enclosed in square brackets like np.array([3]), even though it contains only one element, it's considered a 1-dimensional array. This is because it has one axis along which the single element is placed. So, its dimensionality is 1.

Here's a breakdown:

np.array(3): This creates a zero-dimensional array (scalar array). Dimensionality = 0.
np.array([3]): This creates a one-dimensional array with one element. Dimensionality = 1.
